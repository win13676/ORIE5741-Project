```python
uvals = {}
feature_names = df.columns
for i in range(len(feature_names)):
    uvals[feature_names[i]] = np.unique(df[feature_names[i]][df[feature_names[i]].notna()])
    print(str(i), "\t", str(feature_names[i]),"\t", uvals[feature_names[i]])
```

```python
high_nan = ['Evaporation', 'Sunshine', 'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm']
for name in high_nan:
    med_val = np.mean(uvals[name]) # drop these columns
    df[name] = df[name].fillna(med_val)
```

```python
df = df.dropna()
df
```

```python
dir_labels = ['WindGustDir', 'WindDir9am', 'WindDir3pm']
yn_labels = ['RainToday', 'RainTomorrow']
```

```python
def dir_parser(directions):
    num_north = float(sum([c == 'N' for c in directions]))
    num_south = float(sum([c == 'S' for c in directions]))
    num_east = float(sum([c == 'E' for c in directions]))
    num_west = float(sum([c == 'W' for c in directions]))
    vec = [num_north, num_south, num_east, num_west]
    multiplier = 6 / sum(vec)
    return np.array(vec) * multiplier

def yn_parser(yn):
    return float(yn == 'Yes')
```

```python
df[yn_labels] = df[yn_labels].applymap(yn_parser)

for dirlbl in dir_labels:
    outputs = []
    for v in df[dirlbl]:
        directions = dir_parser(v)
        outputs.append(directions)
    outputs = np.array(outputs)
    df[dirlbl + '_N'] = outputs[:,0]
    df[dirlbl + '_S'] = outputs[:,1]
    df[dirlbl + '_E'] = outputs[:,2]
    df[dirlbl + '_W'] = outputs[:,3]
```

```python
tbdropped = ['Date', 'Location'] + dir_labels
df = df.drop(columns=tbdropped)
df
```

```python
feature_names = df.columns
for i in range(len(feature_names)):
    print(str(i), "\t", str(feature_names[i]),"\t\t\t", str(type(df.iloc[0,i])),"\t",sum(df[feature_names[i]].isna()) / len(df[feature_names[i]]))
```

```python
feature_names = list(feature_names)
feature_names.pop(feature_names.index('RainTomorrow'))
feature_names += ['RainTomorrow']
df = df[feature_names]
```
